# Melt DF

```python
dfc = dfc.melt(id_vars=["Province/State","Country/Region","Lat","Long"],)
```

```python
dfc.rename(columns={"variable":"date","value":"confirmed"},inplace=True)
```

```python
dfc.shape
```

```python
dfc.head(10)
dfc[dfc["date"] == "4/20/21"]
```

```python
dfr = dfr.melt(id_vars=["Province/State","Country/Region","Lat","Long"],)
dfd = dfd.melt(id_vars=["Province/State","Country/Region","Lat","Long"],)
dfr.rename(columns={"variable":"date","value":"recovered"},inplace=True)
dfd.rename(columns={"variable":"date","value":"deaths"},inplace=True)
```

```python
dfc.shape
```

```python
dfd.shape
```

```python
df = pd.merge(dfd,dfc,how='inner',on=["Province/State","Country/Region","Lat","Long","date"])
```

```python
df
```

```python
df1 = df.drop(["deaths","confirmed"],axis=1)
df2 = dfd.drop("deaths",axis=1)
pd.concat([df1,df2]).drop_duplicates(keep=False)
```

```python
dfalgeria = df[df["Country/Region"]== 'Algeria']
```

```python
dfalgeria = dfalgeria.reset_index(drop = True)
dfalgeria["confirmed per day"] = 0
for i in range(dfalgeria.shape[0]-1) :
    dfalgeria.iloc[i,7] = dfalgeria.iloc[i+1,6] - dfalgeria.iloc[i,6]
```

```python
dfalgeria = dfalgeria.reset_index(drop = True)
dfalgeria["deaths per day"] = 0
for i in range(dfalgeria.shape[0]-1) :
    dfalgeria.iloc[i,8] = dfalgeria.iloc[i+1,5] - dfalgeria.iloc[i,5]
```

```python
dfalgeria
```

```python
dfalgeria['date']= pd.to_datetime(dfalgeria['date'])
dfalgeria.dtypes
```

```python
plt.plot(dfalgeria['date'],dfalgeria['confirmed per day'], label='confirmed per day per day')
plt.rcParams["figure.figsize"] = (30,20)
plt.xlabel('date')
plt.ylabel('cases confirmed per day')
plt.show()
```
